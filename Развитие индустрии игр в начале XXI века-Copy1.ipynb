- `year_of_release` имеет тип float64, должен быть int64 - причина может быть в прорусках NaN, которую не поддерживает тип datetime64;
- `eu_sales`, `jp_sales`, `user_score` имеют тип object, должен быть float64 - причина может быть в пропусках, пустых строках или наличии текстовых символов.

## 4. Категоризация данных
    
Проведите категоризацию данных:
- Разделите все игры по оценкам пользователей и выделите такие категории: высокая оценка (от 8 до 10 включительно), средняя оценка (от 3 до 8, не включая правую границу интервала) и низкая оценка (от 0 до 3, не включая правую границу интервала).

```python
def user_score_category(score):
    if pd.isna(score):
        return 'неизвестно'
    elif 8 <= score <= 10:
        return 'высокая'
    elif 3 <= score < 8:
        return 'средняя'
    elif 0 <= score < 3:
        return 'низкая'
    else:
        return 'неизвестно'
```

```python
df_actual['user_score_category'] = df_actual['user_score'].apply(user_score_category)
```

```python
df_actual[['user_score', 'user_score_category']].head(10)
```

- Разделите все игры по оценкам критиков и выделите такие категории: высокая оценка (от 80 до 100 включительно), средняя оценка (от 30 до 80, не включая правую границу интервала) и низкая оценка (от 0 до 30, не включая правую границу интервала).

```python
def critic_score_category(score):
    if pd.isna(score):
        return 'неизвестно'
    elif 80 <= score <= 100:
        return 'высокая'
    elif 30 <= score < 80:
        return 'средняя'
    elif 0 <= score < 30:
        return 'низкая'
    else:
        return 'неизвестно'
```

```python
df_actual['critic_score_category'] = df_actual['critic_score'].apply(critic_score_category)
```

```python
df_actual[['critic_score', 'critic_score_category']].head(10)
```

- После категоризации данных проверьте результат: сгруппируйте данные по выделенным категориям и посчитайте количество игр в каждой категории.

```python
user_cat_counts = df_actual['user_score_category'].value_counts().sort_index()
print("Количество игр по категориям пользовательских оценок:")
print(user_cat_counts)
```

```python
critic_cat_counts = df_actual['critic_score_category'].value_counts().sort_index()
print("Количество игр по категориям оценок критиков:")
print(critic_cat_counts)
```
